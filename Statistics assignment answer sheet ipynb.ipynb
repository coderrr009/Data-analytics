# Statistics Part 2 - Theory Questions

**1. What is hypothesis testing in statistics?**

*Answer:*

Hypothesis testing is a statistical method that uses sample data to evaluate a hypothesis about a population parameter. It helps in making decisions based on data analysis.

**2. What is the null hypothesis, and how does it differ from the alternative hypothesis?**

*Answer:*

The null hypothesis (H0) is a statement of no effect or no difference, and it is tested directly. The alternative hypothesis (H1) represents what we aim to support, indicating some effect or difference.

**3. What is the significance level in hypothesis testing, and why is it important?**

*Answer:*

The significance level (α) is the threshold for rejecting the null hypothesis. Common values are 0.05 or 0.01. It represents the probability of committing a Type I error.

**4. What does a P-value represent in hypothesis testing?**

*Answer:*

A P-value indicates the probability of obtaining the observed results, or more extreme ones, assuming the null hypothesis is true.

**5. How do you interpret the P-value in hypothesis testing?**

*Answer:*

If the P-value is less than the significance level, we reject the null hypothesis. A small P-value suggests strong evidence against H0.

**6. What are Type 1 and Type 2 errors in hypothesis testing?**

*Answer:*

Type I error occurs when a true null hypothesis is rejected. Type II error occurs when a false null hypothesis is not rejected.

**7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**

*Answer:*

A one-tailed test checks for deviation in one direction, while a two-tailed test checks in both directions.

**8. What is the Z-test, and when is it used in hypothesis testing?**

*Answer:*

A Z-test is used when the population standard deviation is known and the sample size is large (n > 30).

**9. How do you calculate the Z-score, and what does it represent in hypothesis testing?**

*Answer:*

Z-score = (X - μ) / σ. It represents how many standard deviations a data point is from the mean.

**10. What is the T-distribution, and when should it be used instead of the normal distribution?**

*Answer:*

The T-distribution is used when the population standard deviation is unknown and the sample size is small (n ≤ 30).

**11. What is the difference between a Z-test and a T-test?**

*Answer:*

Z-test assumes known population standard deviation; T-test does not. T-tests are more versatile for small samples.

**12. What is the T-test, and how is it used in hypothesis testing?**

*Answer:*

A T-test is used to compare means when the sample size is small and population variance is unknown.

**13. What is the relationship between Z-test and T-test in hypothesis testing?**

*Answer:*

Both tests are used to compare means. Z-test is for large samples and known σ, T-test is for small samples and unknown σ.

**14. What is a confidence interval, and how is it used to interpret statistical results?**

*Answer:*

A confidence interval gives a range of values that likely contain the population parameter with a certain level of confidence (e.g., 95%).

**15. What is the margin of error, and how does it affect the confidence interval?**

*Answer:*

Margin of error is the amount added and subtracted from the sample statistic to create a confidence interval. Higher margin = wider interval.

**16. How is Bayes' Theorem used in statistics, and what is its significance?**

*Answer:*

Bayes’ Theorem updates the probability of a hypothesis as more evidence becomes available. It's used in diagnostics, spam filtering, etc.

**17. What is the Chi-square distribution, and when is it used?**

*Answer:*

The Chi-square distribution is used in categorical data analysis, especially for testing independence and goodness-of-fit.

**18. What is the Chi-square goodness of fit test, and how is it applied?**

*Answer:*

The goodness-of-fit test checks whether observed sample data matches expected data from a specific distribution.

**19. What is the F-distribution, and when is it used in hypothesis testing?**

*Answer:*

The F-distribution arises in the comparison of variances. It's used in ANOVA and regression analysis.

**20. What is an ANOVA test, and what are its assumptions?**

*Answer:*

ANOVA (Analysis of Variance) compares means across multiple groups. Assumes independence, normality, and equal variances.

**21. What are the different types of ANOVA tests?**

*Answer:*

Types include One-way ANOVA, Two-way ANOVA, and MANOVA.

**22. What is the F-test, and how does it relate to hypothesis testing?**

*Answer:*

The F-test compares variances of two populations. It's a basis for ANOVA and regression model comparisons.

# Statistics Part 2 - Practical Questions

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
import pandas as pd
```

```python
# 1. Generate a random variable and display its value
random_variable = np.random.rand()
print("1. Random Variable:", random_variable)
```

```python
# 2. Generate a discrete uniform distribution and plot PMF
values = np.arange(1, 7)
p = [1/6] * 6
plt.bar(values, p)
plt.title('2. Discrete Uniform PMF')
plt.show()
```

```python
# 3. PDF of a Bernoulli distribution
def bernoulli_pdf(p, x):
    return p**x * (1-p)**(1-x)
print("3. Bernoulli PDF (p=0.7):", bernoulli_pdf(0.7, 1))
```

```python
# 4. Simulate binomial distribution and plot histogram
binom_data = np.random.binomial(10, 0.5, 1000)
plt.hist(binom_data, bins=10, edgecolor='black')
plt.title('4. Binomial Distribution Histogram')
plt.show()
```

```python
# 5. Create and visualize Poisson distribution
poisson_data = np.random.poisson(3, 1000)
plt.hist(poisson_data, bins=15, edgecolor='black')
plt.title('5. Poisson Distribution')
plt.show()
```

```python
# 6. CDF of discrete uniform distribution
x = np.arange(1, 7)
dist = stats.randint(1, 7)
cdf = dist.cdf(x)
plt.step(x, cdf, where='mid')
plt.title('6. CDF of Discrete Uniform Distribution')
plt.show()
```

```python
# 7. Generate and visualize continuous uniform distribution
uniform_data = np.random.uniform(0, 1, 1000)
plt.hist(uniform_data, bins=20, edgecolor='black')
plt.title('7. Continuous Uniform Distribution')
plt.show()
```

```python
# 8. Simulate data from normal distribution and plot histogram
normal_data = np.random.normal(0, 1, 1000)
plt.hist(normal_data, bins=30, edgecolor='black')
plt.title('8. Normal Distribution Histogram')
plt.show()
```

```python
# 9. Calculate Z-scores and plot them
def calculate_z_scores(data):
    return (data - np.mean(data)) / np.std(data)
z_scores = calculate_z_scores(normal_data)
plt.plot(z_scores)
plt.title('9. Z-scores Plot')
plt.show()
```

```python
# 10. Implement CLT using non-normal distribution
sample_means = [np.mean(np.random.exponential(scale=2, size=30)) for _ in range(1000)]
plt.hist(sample_means, bins=30, edgecolor='black')
plt.title('10. CLT from Exponential Distribution')
plt.show()
```

```python
# 11. Simulate multiple samples and verify CLT
sample_means2 = [np.mean(np.random.normal(0, 1, 30)) for _ in range(1000)]
plt.hist(sample_means2, bins=30, edgecolor='black')
plt.title('11. CLT Verification with Normal Distribution')
plt.show()
```

```python
 #12. Standard normal distribution
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)
plt.plot(x, y)
plt.title('12. Standard Normal Distribution')
plt.show()
```

```python
# 13. Binomial probabilities
bin_probs = stats.binom.pmf(k=np.arange(11), n=10, p=0.5)
plt.bar(np.arange(11), bin_probs)
plt.title('13. Binomial PMF')
plt.show()
```

```python
# 14. Z-score for a data point
x = 75
mean = 70
std = 10
z = (x - mean) / std
print("14. Z-score:", z)
```

```python
# 15. Z-statistics hypothesis testing
sample = np.random.normal(50, 10, 30)
pop_mean = 52
z_stat = (np.mean(sample) - pop_mean) / (np.std(sample)/np.sqrt(len(sample)))
p_val = 2 * (1 - stats.norm.cdf(abs(z_stat)))
print("15. Z-stat:", z_stat, "P-value:", p_val)
```

```python
# 16. Confidence interval
mean_sample = np.mean(sample)
margin_error = stats.norm.ppf(0.975) * (np.std(sample)/np.sqrt(len(sample)))
conf_int = (mean_sample - margin_error, mean_sample + margin_error)
print("16. Confidence Interval:", conf_int)
```

```python
# 17. Confidence interval for normal data
norm_data = np.random.normal(100, 15, 50)
mean = np.mean(norm_data)
sem = stats.sem(norm_data)
conf = stats.t.interval(0.95, len(norm_data)-1, loc=mean, scale=sem)
print("17. CI for Normal Data:", conf)
```

```python
# 18. PDF of normal distribution
x = np.linspace(-4, 4, 100)
y = stats.norm.pdf(x)
plt.plot(x, y)
plt.title('18. Normal PDF')
plt.show()
```

```python
# 19. CDF of Poisson
y = stats.poisson.cdf(np.arange(15), mu=3)
plt.step(np.arange(15), y, where='mid')
plt.title('19. Poisson CDF')
plt.show()
```

```python
# 20. Expected value of continuous uniform distribution
data = np.random.uniform(0, 10, 1000)
expected_value = np.mean(data)
print("20. Expected Value:", expected_value)
```

```python
# 21. Compare standard deviations
data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(0, 2, 100)
print("21. Std Devs:", np.std(data1), np.std(data2))
sns.kdeplot(data1, label='std=1')
sns.kdeplot(data2, label='std=2')
plt.legend()
plt.title('21. Standard Deviation Comparison')
plt.show()
```

```python
# 22. Range and IQR
q75, q25 = np.percentile(normal_data, [75 ,25])
print("22. Range:", np.ptp(normal_data))
print("22. IQR:", q75 - q25)
```

```python
# 23. Z-score normalization
data = np.random.normal(50, 10, 100)
norm_data = (data - np.mean(data)) / np.std(data)
plt.hist(norm_data, bins=20)
plt.title('23. Z-score Normalized Data')
plt.show()
```

```python
# 24. Skewness and kurtosis
print("24. Skewness:", stats.skew(normal_data))
print("24. Kurtosis:", stats.kurtosis(normal_data))
```
